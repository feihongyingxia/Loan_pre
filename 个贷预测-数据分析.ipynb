```python
#过滤数值型类别特征 
def get_numerical_serial_fea(data,feas): 
    numerical_serial_fea = [] 
    numerical_noserial_fea = [] 
    for fea in feas: 
        temp = data[fea].nunique() 
        if temp <= 10: 
            numerical_noserial_fea.append(fea) 
            continue 
        numerical_serial_fea.append(fea) 
    return numerical_serial_fea,numerical_noserial_fea 
numerical_serial_fea,numerical_noserial_fea = get_numerical_serial_fea(data_train,numerical_fea)
print("numerical_fea中的numerical_serial_fea:",numerical_serial_fea)
print("numerical_fea中的numerical_noserial_fea:",numerical_noserial_fea)
```

```python
numerical_serial_fea_internet,numerical_noserial_fea_internet = get_numerical_serial_fea(data_train_internet,numerical_fea_internet)
print("numerical_fea中的numerical_serial_fea:",numerical_serial_fea_internet)
```

```python
numerical_serial_fea_test,numerical_noserial_fea_test = get_numerical_serial_fea(data_test_a,numerical_fea_test)
print("numerical_fea中的numerical_serial_fea:",numerical_serial_fea_test)
```
